```python
import numpy as np

inputs = [1, 2, 3, 2.5]
weights = [
    [0.2, 0.8, -0.5, 1.0],  # weights of the first output neuron
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]
biases = [2, 3, 0.5]

output = np.dot(weights, inputs) + biases
print(output)

```

```python

inputs = [
    [1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]
]

weights = [  # shape (3*4): number of neurons by number of inputs
    [0.2, 0.8, -0.5, 1.0],  # weights of the first output neuron
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]

biases = [2, 3, 0.5]

output = np.dot(inputs, np.array(weights).T) + biases
print(output)
```

```python
import numpy as np

X = [
    [1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]
]

class LayerDense:
    def __init__(self, n_inputs, n_neurons):
        """
        :param n_inputs: number of inputs
        :param n_neurons: numbers of output neurons
        """
        self.weights = 0.10 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases


layer1 = LayerDense(n_inputs=4, n_neurons=5)
layer2 = LayerDense(n_inputs=5, n_neurons=2)
layer1.forward(X)
# print(layer1.output)
layer2.forward(layer1.output)
print(layer2.output)
```
